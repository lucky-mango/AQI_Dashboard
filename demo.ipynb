## Raw data from data.gov of a specific state ##

```python
import requests
import pandas as pd


def get_state_data(state_name: str, api_key: str, limit: int = 1000, offset: int = 0) -> pd.DataFrame:
    """Return only the data records for a selected state as a DataFrame."""
    url = "https://api.data.gov.in/resource/3b01bcb8-0b14-4abf-b6f2-c1bfd384ba69"
    params = {
        "api-key": api_key,
        "format": "json",
        "limit": str(limit),
        "offset": str(offset),
        "filters[state]": state_name,
    }

    response = requests.get(url, params=params)
    if response.status_code != 200:
        raise RuntimeError(f"Failed to fetch data for {state_name}: {response.status_code}")

    data = response.json()
    records = data.get("records", [])
    if not records:
        return pd.DataFrame()

    df = pd.DataFrame.from_records(records)

    # Keep commonly useful columns if they exist
    preferred_columns = [
        "state",
        "city",
        "station",
        "pollutant_id",
        "avg_value",
        "last_update",
        "latitude",
        "longitude",
    ]
    columns_to_keep = [c for c in preferred_columns if c in df.columns]
    if columns_to_keep:
        df = df[columns_to_keep]

    return df


# Example usage: change the state name as needed
selected_state = "Delhi"
API_KEY = globals().get("API_KEY", "579b464db66ec23bdd000001932c55b1958f491a68650e127525c4c6")
state_df = get_state_data(selected_state, API_KEY)

# Display only the data part for the selected state
state_df.head(20)
```

```python
# Improved visualizations replacing crowded box plot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Use existing state_df if present; otherwise fetch a default
try:
    _df = state_df.copy()
except NameError:
    try:
        _df = get_state_data("Delhi", API_KEY)
    except Exception:
        _df = pd.DataFrame()

# Ensure numeric values and drop missing
if not _df.empty and "avg_value" in _df.columns:
    _df = _df.copy()
    _df["avg_value"] = pd.to_numeric(_df["avg_value"], errors="coerce")
    _df = _df.dropna(subset=["avg_value"]) 

sns.set_theme(style="whitegrid", context="notebook")

# 1) Violin plot + jittered points
plt.figure(figsize=(12, 5))
sns.violinplot(data=_df, x="pollutant_id", y="avg_value", inner="quartile", cut=0, scale="width")
sns.stripplot(data=_df, x="pollutant_id", y="avg_value", color="black", alpha=0.25, jitter=0.25, size=2)
plt.title("Distribution of AQI avg_value by pollutant")
plt.xlabel("Pollutant")
plt.ylabel("avg_value")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()

# 2) Horizontal boxen (letter-value) plot
plt.figure(figsize=(12, 5))
sns.boxenplot(data=_df, y="pollutant_id", x="avg_value", orient="h")
plt.title("Letter-value (boxen) plot of avg_value by pollutant")
plt.xlabel("avg_value")
plt.ylabel("Pollutant")
plt.tight_layout()
plt.show()
```
